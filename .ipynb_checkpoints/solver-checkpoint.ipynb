```python
# Tekenafspraak: x naar rechts = positief, y omhoog = positief

def element_stiffness(x1, y1, x2, y2, E, A):
    dx = x2 - x1
    dy = y2 - y1
    L = (dx**2 + dy**2)**0.5

    c = dx / L
    s = dy / L

    import numpy as np
    k = (E * A / L) * np.array([
        [ c*c,  c*s, -c*c, -c*s],
        [ c*s,  s*s, -c*s, -s*s],
        [-c*c, -c*s,  c*c,  c*s],
        [-c*s, -s*s,  c*s,  s*s]
    ])

    return k, L, c, s
```
